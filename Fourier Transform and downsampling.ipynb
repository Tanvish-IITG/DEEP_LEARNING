```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def create_filter(t):
    n = int(t[0])
    m = int(t[1])
    FILTER = np.empty([n,m] , dtype = 'complex128')
    for i in range(n):
        for j in range(m):
            if(i % 3 == 0 and j % 3 == 0):
                FILTER[i,j] = 0
            else:
                FILTER[i,j] = 1
    return FILTER
```

```python
img = cv2.imread("DOG.png",0)      # Reading the image
ft_o = np.fft.fft2(img)            # Taking Fast Fourier Transfrom as given in numpy fft
SPECTRUM_o = np.log(1 + np.abs(ft_o))  # Spectrum is the magnitude of the values in the fast fourier transform
                                       # It is just made to print the spectrum
t = ft_o.shape
FILTER = create_filter(t)
ft_u = FILTER*ft_o                     # create a filter for the input shape and filter the transform
SPECTRUM_u = np.log(1 + np.abs(ft_u))
img_u = np.fft.ifft2(ft_u)             # apply the inverse fourier transform
img_u = np.abs(img_u)                  # print all the images
```

```python
plt.figure(figsize=(6.4*5, 4.8*5), constrained_layout=False)


plt.subplot(151), plt.imshow(img, "gray"), plt.title("Original Image")
plt.subplot(152), plt.imshow(SPECTRUM_o, "gray"), plt.title("Spectrum")
plt.subplot(153), plt.imshow(SPECTRUM_u, "gray"), plt.title("Filtered spectrum")
plt.subplot(154), plt.imshow(img_u, "gray"), plt.title("Final Image")
```
