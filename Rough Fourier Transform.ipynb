```python
n = 427
m = 640
create_filter(n,m):
    FILTER = np.empty([n,m],dtype = 'complex128')
    for i in range(n):
        for j in range(m):
            if(i % 4 == 0):
                FILTER[i,j] = 0
            else:
                FILTER[i,j] = 1
    return FILTER
FILTER = create_filter(n,m)

FILTER
```

```python
img2 = img2 * FILTER
```

```python
img3 = np.fft.ifft2(img2)
```

```python
img3 = np.abs(img3)
```

```python
img1 = cv2.imread("DOG.png",0)
img2 = np.fft.fft2(img1)
spectrum = np.log(1 + np.abs(img2))
img3 = FILTER * img2
Filtered_spectrum = np.log(1 + np.abs(img3))
img4 = np.fft.ifft2(img3)
img4 = np.abs(img4)
```

```python
plt.figure(figsize=(6.4*5, 4.8*5), constrained_layout=False)


plt.subplot(151), plt.imshow(img1, "gray"), plt.title("Original Image")
plt.subplot(152), plt.imshow(spectrum, "gray"), plt.title("Spectrum")
plt.subplot(153), plt.imshow(Filtered_spectrum, "gray"), plt.title("Filtered spectrum")
plt.subplot(154), plt.imshow(img4, "gray"), plt.title("Final Image")
```
